```python
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import statsmodels
from statsmodels.tsa.stattools import coint
import pandas as pd
import scipy.stats as st
import os
import math


pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 100)
```

```python
reload = False

period_ = '2y'

file_name = 'uvxy_' + period_
if not reload and  os.path.exists(file_name):
    hist = pd.read_csv(file_name)
else:
    tmp = yf.Ticker("UVXY")
    hist = tmp.history(period=period_)
    hist.to_csv(file_name)
    
file_name = 'spy_' + period_
if not reload and  os.path.exists(file_name):
    spy_hist = pd.read_csv(file_name)
else:
    tmp = yf.Ticker("SPY")
    spy_hist = tmp.history(period=period_)
    spy_hist.to_csv(file_name)

print(hist.shape)
print(spy_hist.shape)
```

```python
hist.shape
```

```python
# hist.head(10)
# 差分
hist['close_diff']  = hist.Close - hist.Close.shift(1)  + 1000
spy_hist['close_diff']  = spy_hist.Close - spy_hist.Close.shift(1)  + 1000

hist.dropna(inplace=True)
spy_hist.dropna(inplace=True)

# 对数
hist['close_log'] = hist['Close'].apply(math.log)
spy_hist['close_log'] = spy_hist['Close'].apply(math.log)

# 反转
# hist['close_inverse'] = 300-1*hist.Close
hist['close_inverse'] = 1000/hist.Close + 0.5
spy_hist['close_inverse'] = spy_hist.Close
# hist.head()

hist.reset_index(inplace=True)
spy_hist.reset_index(inplace=True)
```

```python
hist['gains'] = (hist.close_inverse -(hist.close_inverse.shift(1)))/(hist.close_inverse.shift(1)) * 100
spy_hist['gains'] = (spy_hist.Close - spy_hist.Close.shift(1))/spy_hist.Close.shift(1) * 100
```

```python
def zscore(series):
    print(series.mean(), series.std())
    return (series - series.mean()) / np.std(series)

z_score = zscore(data.rate)
z_score.plot(figsize=(8,8))
plt.axhline(z_score.mean())
plt.axhline(1.0, color='red')
plt.axhline(-1.0, color='green')
plt.show()
```

```python
# Trade using a simple strategy
def trade(data_, z_score_, window1, window2):
    data_.reset_index(inplace=True,drop=True)
    z_score_.reset_index(inplace=True,drop=True)
    
    S1 = data_['gains_uvxy']
    S2 = data_['gains_spy']
    
    close = data_['Close_uvxy']
    date = data_['Date']
    # If window length is 0, algorithm doesn't make sense, so exit
    if (window1 == 0) or (window2 == 0):
        return 0
    

    money = 0
    countS1 = 0
    countS2 = 0
    gains = []
    # 0 empty position, 1 open position, 2 holding
    status = 0
    
    trade = close.copy()
    
    for i in range(len(S1)):
    
        if status == 0 and  z_score_[i]<-1:
            money += close[i]
            countS1 -= 1    
            status = 1
            print('open position  %s at %f %f %f %s %s'%(date[i],z_score_[i], close[i], money+countS1*close[i], countS1,countS2))
            continue 
        
        if status == 1 and z_score_[i]<-3:
            # stop loss
            money -= close[i]
            countS1 += 1 
            status = 0
            print('stop loss      %s at %f %f %f %s %s'%(date[i],z_score_[i], close[i], money+countS1*close[i], countS1,countS2))
            continue
            
        if status == 1 and z_score_[i] > 1:
            status = 2
            continue
            
        if status == 2 and z_score_[i] < 0:
            status = 0
            money -= close[i]
            countS1 += 1 
            status = 0
            print('clear position %s at %f %f %f %s %s'%(date[i],z_score_[i], close[i], money+countS1*close[i], countS1,countS2))
                
        gains.append(money+countS1*close[i])
   
    gains = np.array(gains)
    sharpe = gains.mean()/gains.std() * np.sqrt(252)
    
    return money+countS1*close.iloc[-1],sharpe

trade(data.iloc[:200],z_score.iloc[:200], 5, 60)
```
